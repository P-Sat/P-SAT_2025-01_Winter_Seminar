```python
#train X25내 결측치 drop
train = train.dropna(subset=['X25'])
```

```python
train['X25'] = train['X25'].apply(lambda x: np.log1p(x))
```

```python
import matplotlib.pyplot as plt
from scipy.stats import shapiro, kstest, probplot
```

```python
# Step 1: Shapiro-Wilk Test
stat_shapiro, p_shapiro = shapiro(train['X25'].dropna())
print(f"Shapiro-Wilk Test: Statistics={stat_shapiro:.3f}, p-value={p_shapiro:.3f}")

# Step 2: Kolmogorov-Smirnov Test
stat_kstest, p_kstest = kstest(train['X25'].dropna(), 'norm', args=(train['X25'].mean(), train['X25'].std()))
print(f"Kolmogorov-Smirnov Test: Statistics={stat_kstest:.3f}, p-value={p_kstest:.3f}")

# Step 3: Q-Q Plot
plt.figure(figsize=(6, 6))
probplot(train['X25'].dropna(), dist="norm", plot=plt)
plt.title("Q-Q Plot for X25")
plt.show()
```

```python
train1 = pd.read_csv('train.csv')
```

```python
train1 = train1.dropna(subset=['X25'])
```

```python
from scipy.stats import boxcox

# 양수 데이터에만 적용 가능
transformed, lambda_value = boxcox(train1['X25'].dropna() + 1e-6)  # 0이 있는 경우 작은 값을 더해줌
print(f"Optimal λ: {lambda_value}")
```

```python
sns.countplot(data=train1, x='X25')
```

```python
original_data = train1['X25'].dropna()

# Shapiro-Wilk Test
stat_orig, p_orig = shapiro(original_data)
stat_trans, p_trans = shapiro(transformed)

print(f"Shapiro-Wilk Test (Original): Statistics={stat_orig:.3f}, p-value={p_orig:.3f}")
print(f"Shapiro-Wilk Test (Transformed): Statistics={stat_trans:.3f}, p-value={p_trans:.3f}")

# Kolmogorov-Smirnov Test
mean_orig, std_orig = original_data.mean(), original_data.std()
mean_trans, std_trans = transformed.mean(), transformed.std()

stat_orig_ks, p_orig_ks = kstest(original_data, 'norm', args=(mean_orig, std_orig))
stat_trans_ks, p_trans_ks = kstest(transformed, 'norm', args=(mean_trans, std_trans))

print(f"Kolmogorov-Smirnov Test (Original): Statistics={stat_orig_ks:.3f}, p-value={p_orig_ks:.3f}")
print(f"Kolmogorov-Smirnov Test (Transformed): Statistics={stat_trans_ks:.3f}, p-value={p_trans_ks:.3f}")
```
